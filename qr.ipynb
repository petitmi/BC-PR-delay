```python
import qrcode

# Basic QR code
url = "https://bc-pr-delay.lol/"
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(url)
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")
img.save("/Users/petitmi/petitmiweb/bc-pr-delay/web_rq.png")
print("QR code saved as web_rq.png")
```
